
**Question 1**

Total multiplication = n\*m\*k  
Total addition = n\*m\*(k-1)

Explanation:
Multipying n\*k and k\*m matrix will result in n\*m matrix i.e the resultant matrix will have n\*m elements. To compute each element R[i, j] in the resultant matrix, we need to perform k multiplication and k-1 addition. This because we multiply each element from the row of matrix A with the corresponding element in the jth column of matrix B, and summing up the products. And since there are n\*m elements in the resultant matrix, total multiplication will be n\*m\*k and total addition will be n\*m\*(k-1).

```python
import jax
import jax.numpy as jnp

def myfunc(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

gradfunc = jax.grad(myfunc, argnums=(0, 1))
x = 2.0
y = 3.0

grad = gradfunc(x, y)

print("Gradient using jax is", grad)
```
